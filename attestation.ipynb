```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.loc[data['whoAmI'] != 'human', 'whoAmI'] = 0
data.loc[data['whoAmI'] == 'human', 'whoAmI'] = 1
```
